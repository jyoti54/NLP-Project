```python
para = " Dr. Kalam also spoke about the importance of facing challenges. He taught students that failure is not the end, but rather an opportunity to learn and grow. He encouraged students to take risks and not be afraid of failure. Don’t take rest after your first victory because if you fail in second, more lips are waiting to say that your first victory was just luck. He reminded students that every challenge they face is a step toward their success.Finally, Dr. Kalam always spoke about the love and respect for one's country. He urged students to contribute to the development of India and make it a better place. He often said, If you want to shine like a sun, first burn like a sun. This means that to achieve greatness, one must be willing to work hard, be dedicated, and face challenges head-on.In conclusion, Dr. A.P.J. Abdul Kalam’s speeches were not just motivational but also practical. He encouraged students to dream, work hard, stay curious, embrace challenges, and contribute positively to society. His life and words continue to inspire students around the world to follow their dreams and make a meaningful impact in the world."
```

```python
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
```

```python
import nltk
nltk.download('stopwords')
```

```python
stopwords.words('english')
```

```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
```

```python
nltk.download('all')
```

```python
nltk.sent_tokenize(para)
```

```python
sentences = nltk.sent_tokenize(para) 
```

```python
type(sentences)
```

```python
##Apply Stopwords and Filter and then apply stemming

for i in range(len(sentences)):
    words = nltk.sent_tokenize(sentences[i])
    words = [stemmer.stem(word) for word in words if word not in set(stopwords.words('english'))]
    sentences[i]=' '.join(words)  ## Converting all the list of words into sentences
```

```python
sentences
```

```python
from nltk.stem import SnowballStemmer
snowballstemmer = SnowballStemmer('english')
```

```python
## snow
for i in range(len(sentences)):
    words = nltk.sent_tokenize(sentences[i])
    words = [snowballstemmer.stem(word) for word in words if word not in set(stopwords.words('english'))]
    sentences[i]=' '.join(words)  ## Converting all the list of words into sentences
```

```python
sentences
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
```

```python
for i in range(len(sentences)):
    words = nltk.sent_tokenize(sentences[i])
    words = [lemmatizer.lemmatize(word,pos='v') for word in words if word not in set(stopwords.words('english'))]
    sentences[i]=' '.join(words)  ## Converting all the list of words into sentences
```
